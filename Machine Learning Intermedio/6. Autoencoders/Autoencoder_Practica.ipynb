<h1 style="*color*:red;font-size:300%;">Aprendizaje no  supervisado.</h1>
<ul style="font-size:200%;line-height:1.5;">
  <li>En problema de aprendizaje supervisionada, contamos como a tupla [x,y].</li>
  <li>En el aprendizaj no supervisado, contamos solamente el vector X. </li>
  <li>En este caso, el objetivo es descobrir alguna cosa respecto a los dados. </li>
  <li>La red tiene que descubrir sola las relaciones, padrones, regularidades o categorias en los datos que le vamos proporcionando y codificandolas en la salida. </li>
  <li>La principal tarea es encontrar la 'mejor' representacion de los datos.</li>
  <li>El entrenamiento no requiere ninguna informacion de clase (labels).</li>
</ul>
<p style="font-size:200%;line-height:2">Modelos no supervisados son mas raros en la practica, pero son muy utiles para direccionar el razonamiento del cientista de datos en el proceso de exploración de los datos para futuros análisis.</p>
</body>

<h1 style="color:red;font-size:300%;">PCA</h1>    
<p style="font-size:200%;line-height:1.5">Técnica utilizada para describir un conjunto de datos en términos de nuevas variables no correlacionadas ("componentes").</p>
<ul style="font-size:200%;line-height:1.5;">
  <li>Toma los datos N-dimensionales y encuentra las dimensiones M ortogonales  en las que los datos tienen la mayor variación</li>
  <li>PCA retiene los componentes con mayor varianza</li>
  <li>La distorsión es mínima entre todos los métodos de reducción de dimensión lineal</li>
  <li>PCA es el algoritmo lineal óptimo para la reducción de dimensionalidad</li>
  </ul>
</body>

<h1 style="color:red;font-size:300%;">Autoencoders</h1>
   
<p style="font-size:200%;line-height:2">Un tipo de redes que están entrenadas para tratar de copiar su entrada a su salida.</p>
    
<ul style="font-size:200%;line-height:1.5;">
  <li>Autoencoder es una forma de realizar la reducción de dimensiones con redes neuronales.</li>
  <li>Autoencoders de varias capas son como una "versión de aprendizaje profundo" del aprendizaje no supervisado.</li>
    <li>Generalmente se usa para disminuir el tamaño de los datos de entrada para facilitar el procesamiento posterior (compactar recursos - features).</li>
  <li>La no linealidad introducida en sus capas puede representar eficientemente datos no lineales.</li>
  <li>Era una idea antigua (que databa de los años 80) pero se consideraba muy difícil de entrenar.</li>
  <li>El primer Autoencoder profundo realmente exitoso fue entrenado en 2006 por el grupo de Hinton.</li>
</ul>
</body>

<body style="background-color:powderblue;">
<h1 style="color:red;font-size:300%;">Estructura Básica de un Autoencoder</h1>

    
<ul style="font-size:200%;line-height:1.5;">
  <li>Como existe un número menor de unidades ocultas, la red es forzada
    en aprender una representacion “compactada” de los datos de entrada.</li>
    <li>La red autoencoder intenta descubrir algunas correlaciones de los datos de entrenamiento.</li>
    <li>Realiza transformaciones en los datos mediante dos tipos de funciones aprendidas.</li>
    <ul style="color:red;">
    <li>Encoder.</li>
    <li>Decoder.</li>
    </ul>
  <li>Necesita de un conjunto de datos de entrenamiento no etiquetados.</li>
  <li>Internamente en la capa oculta, se describe un código usado para representar la entrada.</li>
</ul>

<p style="font-size:200%;line-height:2;color:blue;">Una red autoencoder simple generalmente acaba aprendiendo una representacion muy semejante al algoritmo de PCA.
</p>
</body>    

<body style="background-color:powderblue;">
<h1 style="color:red;font-size:300%;">Encoder</h1>
<p style="font-size:200%;line-height:2;"> Función de extracción de características, que mapea la entrada X en otro espacio llamado espacio
latente (h).
</p>
<p style="font-size:200%;line-height:2;"> $$ 	\mathbf{h}=f(\mathbf{x})=\phi^e (\mathbf{W}^e \mathbf{x} + \mathbf{b}^e) $$ </p>     
</body>   
    

<body style="background-color:powderblue;">
<h1 style="color:red;font-size:300%;">Decoder</h1>
<p style="font-size:200%;line-height:2;">Función de reconstrucción que aprende el mapeo inverso, lleva la representación latente de la capa oculta a la capa de salida (datos reconstruidos).
</p>
<p style="font-size:200%;line-height:2;">$$ \hat{\mathbf{x}}=g(\mathbf{h})=\phi ^d (\mathbf{W}^d \mathbf{h} + \mathbf{b}^d)$$ </p>     
</body>   

    

<body>
<h1 style="font-size:300%;"> Nosotros aprendemos: </h1>

<p style="font-size:200%;line-height:2;">$$ \hat{\mathbf{x}}=g(f(\mathbf{x})) $$</p>     
</body>   





<body style="background-color:powderblue;">

<ul style="font-size:200%;line-height:1.5;">
  <li>No aprenden funciones deterministas (Encoder y Decoder).</li>    
  <li>Aprenden las funciones de reconstrucción y mapeo estocástico.</li>
  <ul style="color:blue;">
      <li>Variational autoencoder (VAE).</li>  
      <li>Adversarial Autoencoder nerwork.</li>        
  </ul>  
</ul>

<h1 style="color:red;font-size:300%;">Variational Autoencoder</h1>
<ul style="font-size:200%;line-height:1.5;">
  <li>Un tipo de autoencoder mas sofisticado.</li>
  <li>Aprende una funcion de distribuicion de la variable latente.</li>
<p>$$ P(\mathbf{h}/ \mathbf{x})$$</p>      
 </body>



http://qytan.com/publication/vae/

```python
from tensorflow.keras import layers
import tensorflow as tf
class Sampling(layers.Layer):
    """Uses (z_mean, z_log_var) to sample z, the vector encoding a digit."""

    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

latent_dim = 64

encoder_inputs = keras.Input(shape=(28, 28, 1))
x = layers.Conv2D(32, 3, activation="relu", strides=2, padding="same")(encoder_inputs)
x = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Flatten()(x)
x = layers.Dense(16, activation="relu")(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
encoder.summary()
latent_inputs = keras.Input(shape=(latent_dim,))
x = layers.Dense(7 * 7 * 64, activation="relu")(latent_inputs)
x = layers.Reshape((7, 7, 64))(x)
x = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(32, 3, activation="relu", strides=2, padding="same")(x)
decoder_outputs = layers.Conv2DTranspose(1, 3, activation="sigmoid", padding="same")(x)
decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()
```

```python
class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(
            name="reconstruction_loss"
        )
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker,
        ]

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)
            reconstruction_loss = tf.reduce_mean(
                tf.reduce_sum(
                    keras.losses.binary_crossentropy(data, reconstruction), axis=(1, 2)
                )
            )
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))
            total_loss = reconstruction_loss + kl_loss
        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }
```

```python
(x_train, _), (x_test, y_test) = keras.datasets.mnist.load_data()
mnist_digits = np.concatenate([x_train, x_test], axis=0)
mnist_digits = np.expand_dims(mnist_digits, -1).astype("float32") / 255

vae = VAE(encoder, decoder)
vae.compile(optimizer='adam')
historyVae =vae.fit(mnist_digits, epochs=30, batch_size=128)
```
